```python
import pandas as pd

df=pd.read_csv("https://raw.githubusercontent.com/soonkienyuan/data-analytics_note/main/dataset/breast_cancer.csv")
```

### d. Generate new features for perimeter_mean and area_mean using log transformation by naming them log_perimeter_mean and log_area_mean. Then drop the older perimeter_mean and area_mean features. Discuss the rationale for doing log transformation.



With the log transformation, the skewness of our original data is reduced or removed, and it is approximately transformed to conform to normality.

Let we take a look the distribution of the original data before log tranformation










```python
#distribution for perimeter_mean and area_mean
import matplotlib.pyplot as plt
%matplotlib inline


df[["perimeter_mean","area_mean"]].hist()
plt.show()

```

Before log transformation, the distributions for perimeter_mean and area_mean are right skewed.

```python
#log transformation
import numpy as np

df[["log_perimeter_mean", "log_area_mean"]] = np.log(df[["perimeter_mean", "area_mean"]])
df.columns
```

```python
# hist after log transform

df[["log_perimeter_mean","log_area_mean"]].hist()
plt.show()
```

After log transformation, the data appears to be approximately normal distributed.

```python
#drop the older perimeter_mean and area_mean features

df.drop(["perimeter_mean","area_mean"], axis=1,inplace= True)
df.head()
```

Answer:

By applying log transformation, we can lower the dataset's variability, which is beneficial for making data patterns more interpretable and for helping to meet the assumptions of inferential statistics.
