Filter contains a specific string from pandas DataFrame

```python
import pandas as pd

df=pd.read_csv("https://raw.githubusercontent.com/soonkienyuan/data-analytics_note/main/dataset/breast_cancer.csv")
```

```python
a= df.filter(regex='_worst|concave|concavity').columns
a
```

```python
#drop all

df.drop(a, axis=1, inplace=True)
```
